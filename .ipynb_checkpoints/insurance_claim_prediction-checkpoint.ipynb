# - Exploratory data analysis

## - Overview


In the train and test data, features that belong to similar groupings are tagged as such in the 
feature names (e.g., ind, reg, car, calc). In addition, feature names include the postfix bin to 
indicate binary features and cat to indicate categorical features. Features without these designations
are either continuous or ordinal. Values of -1 indicate that the feature was missing from the observation. 
The target columns signifies whether or not a claim was filed for that policy holder.





Tags used to indicate features that belong to similar groupings:

 -ind
 
 -reg
 
 -car 
 
 -calc

### Correlation

Analyzing the correlation between features. The objective is to reduce their number by removing highly correlated ones. Highly correlated features may worsen the performance of machine learning models. Furthermore, reducing the number of features improves the speed of training those models.

The analysis will be carried out separately for the categorical, binary, and continuous or ordinal features. The reason is that there are calculations that can be applied to only one type of feature.

```python
#Creating three lists,each containing the names of the categorical features, binary features, and continuous or
#ordinal features

categorical_features  =[i for i in features_list if 'cat' in i] #list of categorical features
binary_features       =[i for i in features_list if 'bin' in i] #list of binary features 
cont_ordinal_features =[i for i in features_list if 'bin' not in i and 'cat' not in i] #list of continuous or ordinal features
cont_ordinal_features.remove('id')
```
